```python
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
from sqlalchemy import text

import pandas as pd
from faker import Faker
from random import randint, choice
from datetime import datetime
```

### Připojení na DB

```python
# base for table classes -> ORM connection between classes and tables
Base = declarative_base()

# connection to DB
engine = create_engine('mysql+pymysql://rektorat:rektorat12345@localhost/projekt_vs')

# link for writing data -> adding and commiting
Session = sessionmaker(bind=engine)
link = Session()
```

### Generování náhodných dat

```python
# generátor studentů
fake = Faker('cs_CZ')
studenti = []

pocet_studentu = 100

for i in range(pocet_studentu):
    # vygeneruj náhodného člověka a jeho zdravotní a platební údaje
    student = {"jmeno": fake.first_name(), "prijmeni": fake.last_name(), "vek": randint(19, 26)}
    studenti.append(student)

studenti_df = pd.DataFrame.from_records(studenti)
studenti_df
```

```python
# generátor učitelů
fake = Faker('cs_CZ')
ucitele = []

pocet_ucitelu = 10

for i in range(pocet_ucitelu):
    # vygeneruj náhodného člověka a jeho zdravotní a platební údaje
    ucitel = {"jmeno": fake.first_name(), "prijmeni": fake.last_name(), "vek": randint(27, 67)}
    ucitele.append(ucitel)

ucitele_df = pd.DataFrame.from_records(ucitele)
ucitele_df
```

```python
# generátor učeben
fake = Faker('cs_CZ')
ucebny = []

pocet_uceben = 4

for i in range(pocet_uceben):
    # vygeneruj náhodného člověka a jeho zdravotní a platební údaje
    ucebna = {"jmeno": fake.first_name_female(), "stat": fake.state(), "mesto": fake.city_name(), "adresa": fake.street_address()}
    ucebny.append(ucebna)

ucebny_df = pd.DataFrame.from_records(ucebny)
ucebny_df
```

```python
# generátor předmětů
fake = Faker('cs_CZ')
predmety = [{"jmeno": "Chemie"}, {"jmeno": "Matematika"}, {"jmeno": "Data Science"}]

predmety_df = pd.DataFrame.from_records(predmety)
predmety_df
```

```python
# generátor zkoušek

datum = datetime.utcnow().isoformat()

zkousky = []
pocet_zkousek = 10

for i in range(pocet_zkousek):
    # vygeneruj náhodného člověka a jeho zdravotní a platební údaje
    zkouska = {"datum": datum, "Ucebna_id": choice(ucebny_df.index + 1), "Student_id": choice(studenti_df.index + 1), "Ucitel_id": choice(ucitele_df.index + 1), "Predmet_id": choice(predmety_df.index + 1)}
    zkousky.append(zkouska)

zkousky_df = pd.DataFrame.from_records(zkousky)
zkousky_df
```

```python
# generátor výsledků

vysledky = []

for i in zkousky:
    body = randint(20, 100)
    prosel = 1 if body > 50 else 0
    # vygeneruj náhodného člověka a jeho zdravotní a platební údaje
    vysledek = {"Zkouska_id": choice(zkousky_df.index + 1), "Prosel": prosel, "Pocet_bodu": body}
    vysledky.append(vysledek)

vysledky_df = pd.DataFrame.from_records(vysledky)
vysledky_df
```

### Uložení dat do databáze

```python
# využijeme .to_sql metodu z pandas -> zápis do sql pomocí pandas
studenti_df.to_sql(name='student', con=engine, if_exists='replace')
ucitele_df.to_sql(name='ucitel', con=engine, if_exists='replace')
predmety_df.to_sql(name='predmet', con=engine, if_exists='replace')
ucebny_df.to_sql(name='ucebna', con=engine, if_exists='replace')
```

```python
zkousky_df.to_sql(name='zkouska', con=engine, if_exists='replace')
vysledky_df.to_sql(name='vysledek', con=engine, if_exists='replace')
```
